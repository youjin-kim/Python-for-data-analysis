# Filtering Data

```python
bios.head()
```

```python
bios.tail()
```

```python
bios.info()
```

```python
bios.loc[bios["height_cm"] > 215]  # , ["name", "height_cm"]
```

```python
bios.loc[(bios["height_cm"] > 215) & (bios["born_country"]=="USA")]
```

```python
bios[bios['name'].str.contains("Keith")]
```

```python
bios[bios['name'].str.contains("keith", case=False)]  # 소문자, 대문자 상관 없음
```

```python
bios[bios['name'].str.contains("keith|patrick", case=False)]
```

```python
bios[bios['name'].str.contains("keith|patrick", case=False, regex=False)]  # 해당하는 정확한 데이터가 없기 때문
```

```python
bios[bios['born_country'].isin(["USA", "FRA", "GBR"]) & (bios['name'].str.startswith("Keith"))]
```

```python
bios.query('born_country == "USA" and born_city == "Seattle"')
```
